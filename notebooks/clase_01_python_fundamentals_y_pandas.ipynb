## 2. Control de flujo

Control de flujo significa decidir y repetir.

Las dos herramientas minimas son:
- `if / else` para tomar decisiones
- `for` para recorrer elementos
- `list comprehensions` para construir listas de forma compacta

### Condicionales

Un condicional evalua si una condicion es verdadera o falsa y ejecuta un bloque segun el resultado.

````python
edad = 20

if edad >= 18:
    print('mayor de edad')
else:
    print('menor de edad')
````

### Loops

Un `for` recorre una secuencia elemento por elemento.

````python
numeros = [1, 2, 3, 4, 5]

for n in numeros:
    print(n * 2)
````

### List comprehensions

Una list comprehension permite crear una nueva lista en una sola linea.

Sirve muy bien para transformaciones simples y filtros cortos.

````python
pares = [n for n in numeros if n % 2 == 0]
print(pares)

pares_2 = []
for n in numeros:
    if n % 2 == 0:
        pares_2.append(n)
print(pares_2)
````

## 3. Funciones

Una funcion recibe entradas, ejecuta una tarea y devuelve un resultado.

La estructura general es:
- `def` para definir
- argumentos de entrada
- codigo interno
- `return` para devolver una salida

````python
def porcentaje_cambio(valor_inicial, valor_final):
    cambio = (valor_final - valor_inicial) / valor_inicial * 100
    return cambio

porcentaje_cambio(100, 120)
````

Una buena funcion encapsula una tarea concreta y se puede reutilizar muchas veces.

## 8. Groupby y agregaciones

Este es uno de los conceptos mas importantes de la sesion.

`groupby` permite dividir la tabla por categorias y luego resumir cada grupo con una metrica como promedio, suma o conteo.

### Promedio por ciudad

````python
df.groupby('City')
df.groupby('City')['Total'].mean()
df.groupby('City')['Total'].mean().sort_values(ascending=False)
````

### Suma por linea de producto

````python
df.groupby('Product line')['Total'].sum().sort_values(ascending=False)
````

### Promedio por tipo de cliente

````python
df.groupby('Customer type')[['Total', 'gross income']].mean()
````

### Groupby con mas de una variable

La estructura mental es:
1. agrupar
2. elegir variable
3. resumir

````python
df.groupby(['City', 'Payment'])['Total'].mean()
````

Reset index se utiliza para poder pasarlo a formato tabla y poder almacenarlo en una variable.

````python
resumen = (
    df.groupby(['City', 'Payment'])['Total']
      .mean()
      .reset_index()
)
resumen.head(10)
````

## 8.1 Merge: juntar tablas

Muchas veces la informacion no viene en una sola tabla. `merge` permite combinar dos bases usando una columna en comun.

La estructura mental es:
1. definir la llave comun
2. elegir que tabla va a la izquierda y cual va a la derecha
3. decidir la logica del cruce con `how`


````python
clientes = pd.DataFrame({
    'cliente_id': [1, 2, 3, 4],
    'cliente': ['Ana', 'Bastian', 'Carla', 'Diego'],
    'ciudad': ['Santiago', 'Valparaiso', 'Concepcion', 'La Serena']
})

compras = pd.DataFrame({
    'cliente_id': [1, 1, 2, 5],
    'producto': ['Cafe', 'Te', 'Jugo', 'Agua'],
    'monto': [3500, 2800, 4100, 1900]
})

display(clientes)
display(compras)
````

La llave comun es `cliente_id`.

- `inner`: deja solo coincidencias en ambas tablas
- `left`: deja todo lo de la tabla izquierda y completa lo que encuentre en la derecha
- `right`: deja todo lo de la tabla derecha y completa lo que encuentre en la izquierda
- `outer`: deja todo lo de ambas tablas, aunque no tengan match


````python
merge_inner = compras.merge(clientes, on='cliente_id', how='inner')
merge_left = compras.merge(clientes, on='cliente_id', how='left')
merge_right = compras.merge(clientes, on='cliente_id', how='right')
merge_outer = compras.merge(clientes, on='cliente_id', how='outer', indicator=True)

print('INNER: solo ids presentes en ambas tablas')
display(merge_inner)

print('LEFT: conserva todas las filas de compras')
display(merge_left)

print('RIGHT: conserva todas las filas de clientes')
display(merge_right)

print('OUTER: conserva todo y muestra de donde viene cada fila')
display(merge_outer)
````
